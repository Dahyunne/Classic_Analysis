```python
import openpyxl
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
classics = pd.DataFrame(columns = ['공연제목', '장소'])
```

```python
res = requests.get("http://ticket.interpark.com/TPGoodsList.asp?Ca=Cla&SubCa=ClassicMain&Sort=4&tid4=ClassicMain")
```

```python
soup = BeautifulSoup(res.content,'html.parser')
```

```python
data = soup.select('tbody tr')
```

```python
for item in data:
    name = item.select_one('.RKtxt a')
    loc = item.select_one('.Rkdate a')
    info = [name.get_text(), loc.get_text()]
    classics.loc[len(classics)] = info
```

```python
classics
```

```python
classics.to_excel('classics.xlsx')
```

```python
import googlemaps
```

```python
gmaps_key = "******************************"
gmaps = googlemaps.Client(key = gmaps_key)
```

```python
classics['장소']
```

```python
address = []
lat = []
lng = []

for l in classics['장소']:
    tmp = gmaps.geocode(l, language = 'ko')
    
    if l == '푸르지오 아트홀':
        address.append('서울특별시 중구 을지로 170')
        lat.append(37.56621791483364)
        lng.append(126.99671750010506)
        continue
        
    if l == '꿈의숲_콘서트홀':
        address.append('서울특별시 강북구 번동 월계로 173')
        lat.append(37.62148976286592)
        lng.append(127.03854410485566)
        continue
        
    if l == '모차르트홀':
        address.append('서울시 서초구 명달로 104')
        lat.append(37.48645666046619)
        lng.append(127.00459343636952)
        continue
        
    if l == '플레이캠퍼스':
        address.append('인천광역시 중구 답동로30번길 9')
        lat.append(37.570281737797934)
        lng.append(126.968441011048)
        continue
        
    if not tmp:
        print(l, "is Null!!!")
        address.append('Null')
        lat.append('NaN')
        lng.append('NaN')
    
    else:
        address.append(tmp[0].get('formatted_address'))
    
        tmp_loc = tmp[0].get("geometry")
    
        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])
    
        print(address[-1])
```

```python
classics['주소'] = address
```

```python
classics.drop(classics.loc[classics['주소'] == '대한민국 서울특별시'].index, inplace = True)
```

```python
classics.drop(classics.loc[classics['장소'].str.contains('해당공연')].index, inplace = True)
```

```python
classics.to_excel('classics.xlsx')
```

```python
import folium
```

```python
map = folium.Map(location=[35.8, 128], zoom_start=7)

for n in concertf.index:
    
    if concertf['공연횟수'][n] >50:
        color = 'red'
    elif concertf['공연횟수'][n] >30:
        color = 'orange'
    elif concertf['공연횟수'][n] >15:    
        color = 'green'
    elif concertf['공연횟수'][n] >5:    
        color = 'blue'
    elif concertf['공연횟수'][n] >3:    
        color = 'purple'
    elif concertf['공연횟수'][n] >1:    
        color = 'gray'
    else:
        color = 'white'
    
    
    folium.Marker([concertf['lat'][n], concertf['lng'][n]],icon=folium.Icon(color = color)).add_to(map)
    
map
```
